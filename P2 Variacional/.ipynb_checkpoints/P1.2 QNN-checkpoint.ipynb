### Introducción


En esta práctica, realizaremos la implementación de una red neuronal cuántica la cual permita clasificar vinos en base a 3 categorías distintas, partiendo de 13 características tales como volumen de alcohol o acidez. 

### Preprocesado

Primeramente, es necesario realizar un preprocesado de los datos, de forma que se ajusten a una entrada manipulable por la red, para esto, dividimos las características y etiquetas en conjuntos de entrenamiento y test, para después llamar a la función MinMaxScaler, encargada de normalizar los datos. En este caso, no realizamos one hot enconding para las etiquetas ya que estos modelos no precisan de este tipo de entrada.

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from IPython.display import clear_output
import time
import numpy as np
from qiskit.primitives import Sampler
from qiskit_machine_learning.algorithms.classifiers import VQC

wine_data = load_wine()
features = wine_data.data
labels = wine_data.target
num_features = features.shape[1]


features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.20)

features_train = MinMaxScaler().fit_transform(features_train)

features_test = MinMaxScaler().fit_transform(features_test)
```

### Desarrollo de la QNN


### State preparation

Ahora debemos de preparar el circuíto que nos permitirá realizar la clasificación, el cual permite transformar las entradas en estados cuánticos. Nuestros primeros modelos se basarán en ZZFeatureMap. 

```python

from qiskit.circuit.library import ZZFeatureMap
featureMap = ZZFeatureMap(feature_dimension=13, reps=1)
featureMap.draw("mpl")
```

### Circuito clasificador


Para realizar la clasificación persé, es necesario crear otro circuíto conocido como clasificador. En primera instancia, utilizaremos el circuíto TwoLocal.

```python

from qiskit.circuit.library import TwoLocal
from qiskit.circuit.library import PauliTwoDesign

classifier = TwoLocal(13, ['ry', 'rz'], 'cz', reps=1)
classifier.draw("mpl")


```

### Optimizador 

Finalmente, definimos el algoritmo de optimización para el proceso de aprendizaje. En primera instancia aplicaremos el algoritmo SPSA.

```python

from qiskit.algorithms.optimizers import SPSA

optimizer = SPSA(maxiter=100)
```

Antes de comenzar la ejecución, definimos la funcion callback_graph, que nos permite mostrar en todo momento la 
evolución de la gráfica de aprendizaje de nuestros modelos.

```python

objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)

#Función que define la gráfica a mostrar mientras entrena el modelo

def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()

```

### Ejecución de la QNN

Ahora, definimos el modelo general con la clase VQC, aplicando todos los componentes creados hasta el momento. De seguido, llamamos a la función de entrenamiento, pasándole los parámetros pertenecientes a este conjunto.

```python
#Caso 1: Classifier TwoLocal, Optimizer SPSA


sampler = Sampler()



vqc = VQC(
    sampler=sampler,
    feature_map=featureMap,
    ansatz=classifier,
    optimizer=optimizer,
    callback=callback_graph,
)


objective_func_vals = []
plt.clf()

start = time.time()
vqc.fit(features_train, labels_train)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")


```

Una vez entrenado, llamamos a la función de score para obtener el accuracy final de nuestro modelo. Llamamos a dicha función tanto para el conjunto de entrenamiento como el de test para comparar la diferencia de resultados entre ambos.

```python
train_score_q1 = vqc.score(features_train, labels_train)
test_score_q1 = vqc.score(features_test, labels_test)
 
print(f"Quantum VQC on the training dataset: {train_score_q1:.2f}")
print(f"Quantum VQC on the test dataset:     {test_score_q1:.2f}")
```

Ahora, procederemos a evaluar diversas variaciones de nuestro modelo, con el fin de intentar mejorara los resultados obtenidos.

### Variación extra: Iris dataset

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python
features = iris_data.data
labels = iris_data.target
```

```python
from sklearn.preprocessing import MinMaxScaler
features = MinMaxScaler().fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
from qiskit.utils import algorithm_globals

algorithm_globals.random_seed = 123
train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, train_size=0.8, random_state=algorithm_globals.random_seed
)
```

```python
from qiskit.circuit.library import ZZFeatureMap

num_features = features.shape[1]
feature_map = ZZFeatureMap(feature_dimension=4, reps=1)
```

```python
from qiskit.circuit.library import RealAmplitudes

ansatz = TwoLocal(4, ['ry', 'rz'], 'cz', reps=1)
```

```python

from qiskit.algorithms.optimizers import SPSA
optimizer = SPSA(maxiter=100)
```

```python
from qiskit.primitives import Sampler
sampler = Sampler()
```

```python
from matplotlib import pyplot as plt
from IPython.display import clear_output

objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)


def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Objective function value against iteration")
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

```python
import time
from qiskit_machine_learning.algorithms.classifiers import VQC

vqc_flor = VQC(
    sampler=sampler,
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer,
    callback=callback_graph,
)

# clear objective value history
objective_func_vals = []

start = time.time()
vqc_flor.fit(train_features, train_labels)
elapsed = time.time() - start

print(f"Training time: {round(elapsed)} seconds")
```

* Let’s see how the quantum model performs on the real-life dataset using the `score` method.

```python
train_score_q4 = vqc_flor.score(train_features, train_labels)
test_score_q4 = vqc_flor.score(test_features, test_labels)

print(f"Quantum VQC on the training dataset: {train_score_q4:.2f}")
print(f"Quantum VQC on the test dataset:     {test_score_q4:.2f}")
```
